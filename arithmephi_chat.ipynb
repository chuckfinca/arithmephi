```python
import flash_attn, torch, locale, pandas, time, random, string, num2words
from transformers import AutoModelForCausalLM, AutoTokenizer
from transformers import BitsAndBytesConfig
```

```python
torch.set_default_device("cuda")
model = AutoModelForCausalLM.from_pretrained("chuckfinca/arithmephi", torch_dtype="auto", trust_remote_code=True)
tokenizer = AutoTokenizer.from_pretrained("microsoft/phi-2", trust_remote_code=True)
```

```python
def response(query):
  inputs = tokenizer(query, return_tensors="pt", return_attention_mask=False)
  outputs = model.generate(**inputs, max_length=len(inputs[0]) + 8, pad_token_id=tokenizer.eos_token_id)
  text = tokenizer.batch_decode(outputs)[0]
  return text
```

```python
response("1867345, 1867346, 1867347")
```
