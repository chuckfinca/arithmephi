```python
import promptbench as pb
```

```python
# print all supported models in promptbench
print('All supported models: ')
print(pb.SUPPORTED_MODELS)

# load a model, flan-t5-large, for instance.
model = pb.LLMModel(model='phi-2', max_new_tokens=10, temperature=0.0001)
```

```python
# print all supported datasets in promptbench
print('All supported datasets: ')
print(pb.SUPPORTED_DATASETS)

# load a dataset, sst2, for instance.
# if the dataset is not available locally, it will be downloaded automatically.
# dataset = pb.DatasetLoader.load_dataset("sst2")
dataset = pb.DatasetLoader.load_dataset("mmlu")
# dataset = pb.DatasetLoader.load_dataset("un_multi")
# dataset = pb.DatasetLoader.load_dataset("iwslt2017", ["ar-en", "de-en", "en-ar"])
# dataset = pb.DatasetLoader.load_dataset("math", "algebra__linear_1d")
# dataset = pb.DatasetLoader.load_dataset("bool_logic")
# dataset = pb.DatasetLoader.load_dataset("valid_parenthesesss")

# print the first 5 examples
dataset[:5]
```

```python
# Prompt API supports a list, so you can pass multiple prompts at once.
prompts = pb.Prompt(["Instruct: {prompt}", "Output:"])
```

```python
# https://huggingface.co/spaces/HuggingFaceH4/open_llm_leaderboard/discussions/26
"""
Turn
{
  "input": "Which of the following groups of drugs may cause a dry mouth?",
  "A": "Steroids.",
  "B": "Analgesics.",
  "C": "Antidepressants.",
  "D": "Antifungal agents.",
  "target": "C",

}

into

Find the degree for the given field extension Q(sqrt(2), sqrt(3), sqrt(18)) over Q.
A. 0
B. 4
C. 2
D. 6
Answer:

"""
def prepare_data(data):
  # print(f"data: {data}")
  prepared_data = {}
  prompt = data["input"]
  for key in data.keys():
    if key == "input" or key == "target" or key == "task":
      continue
    prompt += f"\n{key}. " + data[key].replace(".","")
    prompt += f"\nOutput:"

  prepared_data['prompt'] = prompt
  prepared_data['target'] = data['target']
  prepared_data['task'] = data['task']
  return prepared_data
```

```python
from tqdm import tqdm
for prompt in prompts:
    preds = []
    labels = []
    print(len(dataset))
    for data in tqdm(dataset[:5]):
        data = prepare_data(data)
        # print(f"data: {data}")
        # process input
        # for key in data.keys():
        #   value = data[key]
        #   print(f"key: {key}, value: {value}")
        print("---")
        # answer = { "question" : "this is the question", "answer" : "this is the answer"}

        input_text = pb.InputProcess.basic_format(prompt, data)
        # print(input_text)
        label = data['target']
        print(f"prompt: {input_text}")
        print(f"label: {label}")
        raw_pred = model(input_text)
        print(f"raw_pred: {raw_pred}")
        # process output
        pred = pb.OutputProcess.cls(raw_pred)
        print("-")
        preds.append(pred)
        labels.append(label)

    print(preds)
    print(labels)
    # evaluate
    score = pb.Eval.compute_cls_accuracy(preds, labels)
    print(f"{score:.3f}, {prompt}")
```
